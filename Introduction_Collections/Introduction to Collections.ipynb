# Herança e Polimorfismo

```
from abc import ABCMeta, abstractmethod

class Conta(metaclass = ABCMeta): # Classe abstrata não pode ter objeto instanciado
  def __init__(self, codigo):
    self._codigo = codigo
    self._saldo = 0

  def deposita (self, valor):
    self._saldo += valor

  @abstractmethod # Obrigo a todas as classes filhas implementarem o metodo "passa_mes"
  def passa_mes(self):
    pass 

  def __str__(self):
    return f">> Codigo = {self._codigo} Saldo {self._saldo} <<"
```

```
class ContaCorrente(Conta):

  def passa_mes(self):
    self._saldo -= 2

class ContaPoupanca(Conta):

  def passa_mes(self):
    self._saldo *= 1.01
    self._saldo -= 3

class ContaInvestimento(Conta):
  pass
```

```
ContaInvestimento(765)
```

```
conta16 = ContaCorrente(16)
conta16.deposita(1000)
conta16.passa_mes()
print(conta16)
```

```
conta17 = ContaPoupanca(17)
conta17.deposita(1000)
conta17.passa_mes()
print(conta17)
```

```
conta16 = ContaCorrente(16)
conta16.deposita(1000)
conta17 = ContaPoupanca(17)
conta17.deposita(1000)
contas = [conta16, conta17]
for conta in contas:
  conta.passa_mes() # Duck Typing
  print(conta)
```

# __*CHAPTER 4*__ - Metodo __eq__ (Igualdade)

```
class ContaSalario:
  def __init__(self, codigo):
    self._codigo = codigo
    self._saldo = 0

  def __eq__(self, outro): #Função de define o critério de igualdade
    ### Nosso Critério para definir se é igual ou não
    #if type(outro) != ContaSalario:
      #return False

    #
    return self._codigo == outro._codigo and self._saldo == outro._saldo \
        and isinstance(outro, ContaSalario) #"isinstance" verifica se o objeto faz parte da Hierarquia da classe Mãe (informada)

  def deposita (self, valor):
    self._saldo += valor

  def __str__(self):
    return f">> Codigo = {self._codigo} Saldo {self._saldo} <<"

class ContaMultiSalario(ContaSalario):
  pass
```

```
conta1 = ContaSalario(31)
print(conta1)
conta2 = ContaCorrente(31) # Mesmo código que "conta1"
print(conta2)
conta1 == conta2
```

```
conta1 != conta2 # "!=" Diferente também é baseado no __eq__
```

```
conta1 in [conta2]
```

```
conta1.deposita(1000)
conta1 == conta2
```

```
conta1 = ContaSalario(31)
print(conta1)
conta2 = ContaMultiSalario(31) #Faz parte da Hierarquida da ContaSalario
print(conta2)
conta1 == conta2
```

# __*CHAPTER 7*__ - Custom Order

```
class ContaSalario:
  def __init__(self, codigo):
    self._codigo = codigo
    self._saldo = 0

  def __eq__(self, outro):
    ### Nosso Critério para definir se é igual ou não
    if type(outro) != ContaSalario:
      return False
    return self._codigo == outro._codigo and self._saldo == outro._saldo

# Função criada para comparar se um objeto do tipo ContaSalario é menor que (less than) outra ContaSalario
  def __lt__(self, outro):
    return self._saldo < outro._saldo

  def deposita (self, valor):
    self._saldo += valor

  def __str__(self):
    return f">> Codigo = {self._codigo} Saldo {self._saldo} <<"

class ContaMultiSalario(ContaSalario):
  pass
```

```
conta_levy = ContaSalario(17)
conta_levy.deposita(500)
conta_marina = ContaSalario(3)
conta_marina.deposita(1000)
conta_paulo = ContaSalario(133)
conta_paulo.deposita(510)

contas = [conta_levy, conta_marina, conta_paulo]
```

```
for conta in contas:
  print(conta)
```

```
sorted(contas)
```

```
def extrai_saldo(conta):
  return conta._saldo # Acesso indevido em atributo com "_"

for conta in sorted(contas, key=extrai_saldo):
# Parm "key" define o atributo utilizado para ordenar os objetos da Lista (contas)
  print(conta)
```

```
from operator import attrgetter #Função que faz acesso a um atributo de qlqr objeto
# Não precisa criar a função (extrai_saldo)
for conta in sorted(contas, key=attrgetter('_saldo')): # Continua acessando indevidamente um atributo com '_'
  print(conta)
```

```
# Funciona após a implementação da função "lt" na classe ContaSalario
print(conta_marina < conta_levy)
print( conta_marina > conta_levy )
```

```
for conta in sorted(contas): # Não precisa passar o parm "key" após a função "lt" na classe
  print(conta)
print("\nReverse=True\n")
for conta in sorted(contas, reverse=True):
  print(conta)
```

```
class ContaSalario:
  def __init__(self, codigo):
    self._codigo = codigo
    self._saldo = 0

  def __eq__(self, outro):
    if type(outro) != ContaSalario:
      return False
    return self._codigo == outro._codigo and self._saldo == outro._saldo

# Função criada para comparar se um objeto do tipo ContaSalario é menor que (less than) outra ContaSalario
  def __lt__(self, outro):
    if self._saldo != outro._saldo:
      return self._saldo < outro._saldo
    return self._codigo < outro._codigo


  def deposita (self, valor):
    self._saldo += valor

  def __str__(self):
    return f">> Codigo = {self._codigo} Saldo {self._saldo} <<"

class ContaMultiSalario(ContaSalario):
  pass
```

```
# total_ordering faz a Ordenação Total (ver documentação)
from functools import total_ordering

@total_ordering
class ContaSalario:
  def __init__(self, codigo):
    self._codigo = codigo
    self._saldo = 0

  def __eq__(self, outro):
    if type(outro) != ContaSalario:
      return False
    return self._codigo == outro._codigo and self._saldo == outro._saldo

# Função criada para comparar se um objeto do tipo ContaSalario é menor que (less than) outra ContaSalario
  def __lt__(self, outro):
    if self._saldo != outro._saldo:
      return self._saldo < outro._saldo
    return self._codigo < outro._codigo


  def deposita (self, valor):
    self._saldo += valor

  def __str__(self):
    return f">> Codigo = {self._codigo} Saldo {self._saldo} <<"

class ContaMultiSalario(ContaSalario):
  pass
```

```
# Funciona após a implementação do "total_ordering"
print(conta_levy <= conta_marina)
print(conta_marina <= conta_levy)
print(conta_levy >= conta_marina)
print(conta_levy < conta_levy)
print(conta_levy == conta_levy)
print(conta_levy <= conta_levy)
```
